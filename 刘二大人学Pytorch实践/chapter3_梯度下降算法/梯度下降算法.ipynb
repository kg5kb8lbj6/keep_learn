### 1. 梯度下降

```python
# y = 4 * w
```

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
w = 1.0
```

```python
# 前向传播
def forward(x):
    return x * w

# 求总的损失值
def cost_fun(x_data, y_data):
    cost = 0 
    for x,y in zip(x_data, y_data):
        y_pred = forward(x)
        cost += (y_pred - y) ** 2
    return cost / len(x_data)

def gradient_fun(x_data, y_data):
    grad = 0
    for x, y in zip(x_data, y_data):
        grad += 2 *(x * w - y) * x
    return grad / len(x_data)
```

```python
cost_list = []
epoch_list = []
for epoch in range(100):
    cost_val = cost_fun(x_data, y_data)
    grad_val = gradient_fun(x_data, y_data)
    w -= 0.01 * grad_val
    print('Epoch:', epoch, 'w = ', w, 'loss = ', cost_val)
    cost_list.append(cost_val)
    epoch_list.append(epoch)
print('predict (before training)', 4, forward(4))
```

```python
plt.plot(epoch_list, cost_list)
plt.xlabel('Eopch')
plt.ylabel('Loss', rotation = 0)
```

### 2. 随机梯度下降

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
w = 1.0
```

```python
def forward(x):
    return x * w

def loss_fun(x, y):
    y_pred = forward(x)
    return (y_pred - y) ** 2

def gradient(x, y):
    return 2 * (x * w - y) * x
```

```python
loss_list = []
epoch_list = []
for epoch in range(100):
    for x, y in zip(x_data, y_data):
        grad = gradient(x, y)
        w -= 0.01 * grad
        print('\grad:', x, y, grad)
        l = loss_fun(x ,y)
    loss_list.append(l)
    epoch_list.append(epoch)
    print('Epoch: ', epoch, 'w = ', w, 'loss = ', l)
print('predict (before training)', 4, forward(4))
```

```python
plt.plot(epoch_list, loss_list)
plt.xlabel('Epoch')
plt.ylabel('Loss', rotation = 0)
plt.show()
```
