```python
import torch
import matplotlib.pyplot as plt
```

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
```

```python
w = torch.tensor([1.0], requires_grad = True)
w
```

```python
def forward(x):
    return x * w

def loss_fun(x, y):
    y_pred = forward(x)
    return (y_pred - y) ** 2
```

```python
for epoch in range(100):
    print('Epoch =', epoch)
    for x, y in zip(x_data, y_data):
        loss = loss_fun(x, y)
        print('\tx = {}, y = {}, loss = {}'.format(x, y, loss))
        loss.backward()
        w.data = w.data - 0.01 * w.grad.data
        w.grad.data.zero_()
print('forward(4) =  ', forward(4))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import torch

x_data = [1.0,2.0,3.0]
y_data = [2.0,4.0,6.0]
```

```python
w1 = torch.tensor([1.0], requires_grad = True)
w2 = torch.tensor([1.0], requires_grad = True)
b = torch.tensor([1.0], requires_grad = True)
```

```python
def forward(x):
    f = w1 * x ** 2 + w2 * x + b
    return f
```

```python
def loss_fun(x, y):
    y_hat = forward(x)
    loss = (y_hat - y) ** 2
    return loss
```

```python
lr = 0.01
for epoch in range(100): # 1
    print('Epoch = ', epoch)
    for x, y in zip(x_data, y_data):
        #print(x, y)
        loss = loss_fun(x, y)
        loss.backward()
        # 更新参数
        print('\tw1 = {}, w2 = {}, b = {}'.format(w1.data, w2.data, b.data))
        w1.data = w1.data - lr * w1.grad.data
        w2.data = w2.data - lr * w2.grad.data
        b.data = b.data - lr * b.grad.data
        # 梯度清零
        w1.grad.data.zero_()
        w2.grad.data.zero_()
        b.grad.data.zero_()
```
